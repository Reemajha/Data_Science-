```python
##Quoting and Escape Characters very useful in NLP (Natural Language Processing)
# e.g. In Gmail, there are some emails that aumatically goes into Promotion, some goes in Spam, some comes in Social. How? 
# This is done with the help of NLP 
```

```python
data = 'a,b\n"hello, \\"Bob\\", nice to see you",5' # \\ is escape character 
```

```python
data
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
